```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns


round1_df = pd.read_csv("../../data/round1.csv", delimiter= ";")
round1_df.head()
```

```python
ink_df = round1_df[round1_df["product"] == "SQUID_INK"]
ink_df.head()
```

```python
sns.scatterplot(data = ink_df, x = "timestamp", y = "bid_price_1", label = "bid price 1")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
sns.scatterplot(data = ink_df, x = "timestamp", y = "bid_price_2", label = "bid price 2")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
sns.scatterplot(data = ink_df, x = "timestamp", y = "bid_price_3", label = "bid price 3")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
sns.scatterplot(data = ink_df, x = "timestamp", y = "ask_price_1", label = "ask price 1")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
sns.scatterplot(data = ink_df, x = "timestamp", y = "ask_price_2", label = "ask price 2")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
sns.scatterplot(data = ink_df, x = "timestamp", y = "ask_price_3", label = "ask price 3")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
ink_df["mean_ask_price"] = ink_df[["ask_price_1", "ask_price_2", "ask_price_3"]].mean(axis = 1)
sns.scatterplot(data = ink_df, x = "timestamp", y = "mean_ask_price", label = "mean_ask_price")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
ink_df["bid_diff"] = ink_df["bid_price_1"] - ink_df["bid_price_2"]
sns.scatterplot(data = ink_df, x = "timestamp", y = "bid_diff", label = "bid_diff 1 v 2")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
ink_df["bid_diff"] = ink_df["bid_price_2"] - ink_df["bid_price_3"]
sns.scatterplot(data = ink_df, x = "timestamp", y = "bid_diff", label = "bid_diff 2 v 3")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
ink_df["ask_diff"] = ink_df["ask_price_1"] - ink_df["ask_price_2"]
sns.scatterplot(data = ink_df, x = "timestamp", y = "bid_diff", label = "ask_diff 1 v 2")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
ink_df["ask_diff"] = ink_df["ask_price_2"] - ink_df["ask_price_3"]
sns.scatterplot(data = ink_df, x = "timestamp", y = "bid_diff", label = "ask_diff 2 v 3")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```
