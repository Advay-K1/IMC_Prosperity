```python
from pandas.plotting import autocorrelation_plot

ink_df['timestamp'] = pd.to_datetime(ink_df['timestamp'])

# --- 1. Mid Price Over Time ---
plt.figure(figsize=(12, 4))
plt.plot(ink_df['timestamp'], ink_df['mid_price'], label='Mid Price')
plt.title("SQUID_INK Mid Price Over Time")
plt.xlabel("Time")
plt.ylabel("Mid Price")
plt.grid()
plt.legend()
plt.show()

# --- 2. Price Change Distribution ---
ink_df['price_change'] = ink_df['mid_price'].diff()
plt.figure(figsize=(8, 4))
ink_df['price_change'].hist(bins=100)
plt.title("Histogram of Tick-to-Tick Price Changes")
plt.xlabel("Price Change")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# --- 3. Rolling Volatility (50-tick window) ---
ink_df['rolling_std_20'] = ink_df['mid_price'].rolling(20).std()
plt.figure(figsize=(10, 4))
plt.plot(ink_df['rolling_std_20'], label='Rolling Std (20)')
plt.title("SQUID_INK Rolling Volatility (20 ticks)")
plt.xlabel("Tick")
plt.ylabel("Volatility")
plt.grid()
plt.legend()
plt.show()

# --- 4. Deviation from 2000 Anchor ---
ink_df['dev_from_2000'] = ink_df['mid_price'] - 2000
plt.figure(figsize=(8, 4))
ink_df['dev_from_2000'].hist(bins=100)
plt.title("Deviation from Static Anchor (2000)")
plt.xlabel("Mid Price - 2000")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# --- 5. Deviation from 50 Ticks Ago (Reversion Potential) ---
ink_df['anchor_50'] = ink_df['mid_price'].shift(50)
ink_df['deviation_50'] = ink_df['mid_price'] - ink_df['anchor_50']
plt.figure(figsize=(8, 4))
ink_df['deviation_50'].hist(bins=100)
plt.title("Deviation from Price 50 Ticks Ago")
plt.xlabel("Price - Price[50 ticks ago]")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# --- 6. Autocorrelation of Price Changes ---
plt.figure(figsize=(8, 4))
autocorrelation_plot(ink_df['price_change'].dropna())
plt.title("Autocorrelation of Price Changes")
plt.grid()
plt.show()
```

```python
# === 1. Deviation from 50-tick anchor ===
ink_df['anchor_50'] = ink_df['mid_price'].shift(50)
ink_df['deviation_50'] = ink_df['mid_price'] - ink_df['anchor_50']

plt.figure(figsize=(8, 4))
ink_df['deviation_50'].hist(bins=100)
plt.title("Deviation from Price 50 Ticks Ago")
plt.xlabel("Deviation")
plt.ylabel("Frequency")
plt.grid()
plt.show()

# === 2. Time-to-Revert from Big Deviations ===
threshold = 50
revert_zone = 10
max_lookahead = 100

reversion_times = []
reversion_success = []

for i in range(len(ink_df) - max_lookahead):
    anchor = ink_df.iloc[i]['mid_price']
    reverted = False
    for j in range(i + 1, i + max_lookahead):
        if abs(ink_df.iloc[j]['mid_price'] - anchor) < revert_zone:
            reversion_times.append(j - i)
            reversion_success.append(True)
            reverted = True
            break
    if not reverted:
        reversion_times.append(max_lookahead)
        reversion_success.append(False)

# === 3. Naive Reversion Strategy Simulation ===
positions = []
pnl = []
holding = None
entry_price = None
exit_zone = 10

for i in range(len(ink_df)):
    if i < 50:
        positions.append(0)
        pnl.append(0)
        continue

    anchor = ink_df.iloc[i - 50][ 'mid_price']
    price = ink_df.iloc[i][ 'mid_price']

    if holding is None:
        if price > anchor + threshold:
            holding = -1  # short
            entry_price = price
        elif price < anchor - threshold:
            holding = 1  # long
            entry_price = price
        positions.append(0)
        pnl.append(0)
    else:
        if abs(price - anchor) < exit_zone:
            profit = (entry_price - price) * holding
            pnl.append(profit)
            positions.append(0)
            holding = None
            entry_price = None
        else:
            pnl.append(0)
            positions.append(holding)

# === 4. Volatility Regime Split ===
ink_df['rolling_std_20'] = ink_df['mid_price'].rolling(20).std()
low_vol = ink_df[ink_df['rolling_std_20'] < 2]
high_vol = ink_df[ink_df['rolling_std_20'] > 4]

# === 5. Print Summary ===
print(" Reversion Test Summary")
print(f"Total swings tested:         {len(reversion_times)}")
print(f"Successful reversions:       {sum(reversion_success)}")
print(f"Reversion rate:              {np.mean(reversion_success):.2%}")
print(f"Avg time to revert (success): {np.mean([t for t, s in zip(reversion_times, reversion_success) if s]):.2f}")

print("\n Naive Strategy Simulation")
print(f"Total trades executed:       {len([p for p in pnl if p != 0])}")
print(f"Total PnL:                   {sum(pnl):.2f}")
print(f"Avg PnL per trade:           {np.mean([p for p in pnl if p != 0]):.2f}")

print("\n Volatility Regimes")
print(f"Low-volatility samples:      {len(low_vol)}")
print(f"High-volatility samples:     {len(high_vol)}")
```
