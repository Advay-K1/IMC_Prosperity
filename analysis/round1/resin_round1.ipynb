```python
"""
Proportion-based buy-sell logic
"""
def rr_strategy1(rr_df):
    rr_df_size = len(rr_df)
    mean_price = rr_df["mid_price"].mean()

    # These proportions are currently unused, but you might use them later
    above_mean = len(rr_df[rr_df["mid_price"] > mean_price])
    below_mean = len(rr_df[rr_df["mid_price"] < mean_price])

    ratio_above_mean = above_mean / rr_df_size
    ratio_below_mean = below_mean / rr_df_size

    # Set thresholds based on the mean
    sell_threshold = mean_price + 1
    buy_threshold = mean_price - 1.5

    return sell_threshold, buy_threshold

def test_strategy1(rr_df_test):
    profit = 0
    position = 0 
    position_limit = 50
    position_arr, profit_arr = [], []

    sell_threshold, buy_threshold = rr_strategy1(rr_df_test)

    for i in range(len(rr_df_test)):
        row = rr_df_test.iloc[i]
        mid_price = row["mid_price"]

        if mid_price >= sell_threshold:
            if position - row["bid_volume_1"] >= -position_limit:
                position -= row["bid_volume_1"]
                profit +=  row["bid_price_1"]


        elif mid_price <= buy_threshold:
            if position + row["ask_volume_1"] <= position_limit:
                position += row["ask_volume_1"]
                profit -= row["ask_price_1"]


        profit_arr.append(profit)
        position_arr.append(position)

        print(f"iteration: {i}, position: {position}, profit: {profit}")

    return position_arr, profit_arr

position_arr, profit_arr = test_strategy1(rr_df)
```

```python
df = pd.DataFrame({'data': profit_arr})

sns.lineplot(x=df.index, y='data', data=df, label = "profit")
plt.show()
```

```python
df = pd.DataFrame({'data': position_arr})

sns.lineplot(x=df.index, y='data', data=df, label = "position")
plt.show()
```
