```python
kelp_df = kelp_df.copy()
kelp_df['mid_price'] = (kelp_df['bid_price_1'] + kelp_df['ask_price_1']) / 2
kelp_df['spread'] = kelp_df['ask_price_1'] - kelp_df['bid_price_1'] - 0.6


kelp_df['sigma'] = kelp_df['mid_price'].rolling(window=10).std()
kelp_df['sigma_sq'] = kelp_df['sigma'] ** 2


avg_spread = kelp_df['spread'].dropna().mean()
avg_sigma_sq = kelp_df['sigma_sq'].dropna().mean()

print(f"Observed average spread: {avg_spread:.2f}")
print(f"Estimated average sigma: {avg_sigma_sq:.6f}")


gamma_vals = np.linspace(0.00008, 0.00012, 100)
kappa_vals = np.linspace(0.26, 0.35, 100)

total_ticks = 20000
avg_tick = kelp_df['timestamp'].mean()
time_rem = total_ticks - avg_tick

def stokoviv_spread(gamma, kappa, sigma_sq, time_rem):
    return gamma * sigma_sq * time_rem + (2 / gamma) * np.log(1 + gamma / kappa)

results = []
for gamma in gamma_vals:
    for kappa in kappa_vals:
        spread = stokoviv_spread(gamma, kappa, avg_sigma_sq, time_rem)
        if abs(spread - avg_spread) < 0.005:  
            results.append((gamma, kappa, spread))

print(f"\nParameter combinations that match target spread  {avg_spread:.2f}:\n")
for gamma, kappa, spread in results:
    print(f"  ={gamma:.4f}, ={kappa:.3f}  predicted spread={spread:.2f}")
```
