```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns


round0_df = pd.read_csv("../data/BoilerPlateCode.csv", delimiter= ";")
round0_df.head()
```

```python
rr_df = round0_df[round0_df["product"] == "RAINFOREST_RESIN"]
rr_df.head()
```

```python
sns.scatterplot(data = rr_df, x = "timestamp", y = "bid_price_1", label = "bid price 1")
sns.scatterplot(data = rr_df, x = "timestamp", y = "bid_price_2", label = "bid price 2")
sns.scatterplot(data = rr_df, x = "timestamp", y = "bid_price_3", label = "bid price 3")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
sns.scatterplot(data = rr_df, x = "timestamp", y = "ask_price_1", label = "ask price 1")
sns.scatterplot(data = rr_df, x = "timestamp", y = "ask_price_2", label = "ask price 2")
sns.scatterplot(data = rr_df, x = "timestamp", y = "ask_price_3", label = "ask price 3")
plt.xlabel("Time Stamp")
plt.ylabel("Sea Shells")
```

```python
rr_df["mean_bid_price"] = rr_df[["bid_price_1", "bid_price_2", "bid_price_3"]].mean(axis = 1)
rr_df["mean_ask_price"] = rr_df[["ask_price_1", "ask_price_2", "ask_price_3"]].mean(axis = 1)
sns.scatterplot(data = rr_df, x = "timestamp", y = "mean_bid_price", label = "average of bid price iteration")
sns.scatterplot(data = rr_df, x = "timestamp", y = "mean_ask_price", label = "average of ask price iteration")
```

```python
rr_df["mean_bid_price"].describe()
```

```python
rr_df["mean_ask_price"].describe()
```

```python
#difference of mean ask and mean bid prices

rr_df["mean_bid_ask_diff"] = rr_df["mean_bid_price"] - rr_df["mean_ask_price"]
rr_df.head()
```

```python
print (rr_df["mean_bid_ask_diff"].describe())
sns.scatterplot(data = rr_df, x = "timestamp", y = "mean_bid_ask_diff", label = "average of difference of bid and ask price iteration")
```
