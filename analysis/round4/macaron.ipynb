```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns


df = pd.read_csv("../../data/round_4_day_1_intermediaries.csv", sep=",")
macaron_df = pd.read_csv("../../data/round4.csv", delimiter= ";")

skip = {'timestamp', 'bidPrice', 'askPrice'}

for col in df.columns:
    if col in skip:
        continue
    plt.figure()
    plt.plot(df['timestamp'], df[col])
    plt.xlabel('Timestamp')
    plt.ylabel(col)
    plt.title(f'{col} Over Time')
    plt.tight_layout()
    plt.show()

macarons = macaron_df[macaron_df["product"] == "MAGNIFICENT_MACARONS"]

plt.figure()
plt.plot(macarons["timestamp"], macarons["mid_price"], marker="o", linestyle="-")
plt.xlabel("Timestamp")
plt.ylabel("Mid Price")
plt.title("MAGNIFICENT_MACARONS Mid Price Over Time")
plt.grid(True)
plt.tight_layout()
plt.show()
```
